Quick over view of clustering methods.

Go over this notebook to see how to use sklearn for clustering.

```python
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
```

```python
iris_data= datasets.load_iris()
```

```python
X = iris_data.data
y= iris_data.target
iris_data.target_names
```

```python
X.shape, y.shape
```

Let's reduce dimension to two

```python
kpca = PCA(n_components=2)
```

```python
X_kpca = kpca.fit_transform(X)
```

```python
kmean = KMeans(n_clusters=3, random_state=1)
```

```python
kmean.fit(X_kpca)
```

```python
kmean.labels_.tolist()
```

**Colour show true label and number is learned cluster label**

```python
f, ax = plt.subplots(figsize=(10,10))
ax.scatter(X_kpca[:,0], X_kpca[:,1], c= y)
for i, lbl in enumerate(kmean.labels_.tolist()):
    ax.annotate(lbl, (X_kpca[i,0], X_kpca[i, 1]))
```

Look at the documentation for other linkage methods

```python
aggl_cltsering = AgglomerativeClustering(n_clusters=3, linkage='ward')
```

```python
aggl_cltsering.fit(X_kpca)
```

```python
f, ax = plt.subplots(figsize=(10,10))
ax.scatter(X_kpca[:,0], X_kpca[:,1], c= y)
for i, lbl in enumerate(aggl_cltsering.labels_.tolist()):
    ax.annotate(lbl, (X_kpca[i,0], X_kpca[i, 1]))
```

Dendograms

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linkage_info = linkage(X_kpca, method='ward')
```

```python
plt.figure(figsize=(20,20))
dendrogram(linkage_info)
plt.ylabel('distance')
plt.xlabel('sample id')
```
