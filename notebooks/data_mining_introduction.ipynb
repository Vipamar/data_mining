# Let's plot first two attribute of this dataset

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5



# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
X = iris.data[:, :3]  # we only take the first two features.
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
#X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three attribute")
ax.set_xlabel("Septal length")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("Septal Width")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("Petal Length")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

How to visualize data bigger than 3 dim

```python
from sklearn.decomposition import PCA
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```

## Dot product of two vector $v_1, v_2 \in \mathbb{R}^d$

$v_1=  \begin{bmatrix} v_{11}\\ v_{12}\\ \vdots\\ v_{1d}  \end{bmatrix}$ and $v_2=  \begin{bmatrix} v_{21}\\ v_{22}\\ \vdots\\ v_{2d}  \end{bmatrix}$


$v_1^Tv_2 = \sum_i^d v_{1i}v_{2i}$

```python
np.dot(v_1, v_2)
```

# Length of a vector
$ \|v_1 \|_2 = v_1^Tv_1 = (\sum_i^d v_{1i}v_{1i})^{1/2}$

```python
np.sqrt(np.dot(v_1, v_1))
```

or use linear algebra module from numpy

```python
np.linalg.norm(v_1, ord=2)
```

## Distance between two vector v_1, v_2
$\|v_1 - v_2 \|_2$

```python
np.linalg.norm(v_1- v_2, ord=2)
```
